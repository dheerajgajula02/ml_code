```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('50_Startups.csv')
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct= ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[3])], remainder='passthrough')
x=ct.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.2, random_state=0)
x_train
```

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()

regression.fit(x_train,y_train)
```

```python
y_predict= regression.predict(x_train)
np.set_printoptions(precision=3)
```

```python
print(np.concatenate((y_train.reshape(len(y_train),1), y_predict.reshape(len(y_predict),1)),1))
```

```python
print(x_test)
```

```python
plt.scatter(x_train[:,3],y_train, edgecolors='red')
plt.plot(x_train[:,3],y_predict)
plt.xlabel("R & D spent")
plt.ylabel("profit")
plt.title("R&D vs profit")
plt.show()
```

```python
plt.scatter(x_train[:, 4], y_train, edgecolors='red')
plt.plot(x_train[:, 4], y_predict)
plt.xlabel("administration spent")
plt.ylabel("profit")
plt.title("admin vs profit")
plt.show()
```
