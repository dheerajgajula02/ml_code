logistic regression is used to predict yes or not ( kinda )

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
(x_test)
```

```python
print(y_test)
x_test
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
x_train=scale.fit_transform(x_train)
x_test=scale.transform(x_test)
```

```python

(x_train)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=0)
classifier.fit(x_train,y_train)
```

```python
classifier.predict(scale.transform([x[0]]))
```

```python
classifier.predict(([x_train[0]]))
```

```python
classifier.predict([x_test[0]])
```

```python
classifier.predict_proba([x_test[0]])
```

```python
y_predict=classifier.predict(x_test)
```

```python
print(y_predict)
```

```python
print(y_test)
```

```python
np.concatenate((scale.inverse_transform(x_test)  , y_predict.reshape(len(y_predict),1),y_test.reshape(len(y_test),1)),1)
```

confusion matrix basically tells us which are correct
and which are wrong predictions

```python
from sklearn.metrics import confusion_matrix, accuracy_score

(confusion_matrix(y_test, y_predict))
```

confusion matrix's nomenclature
[ correct prediction of people who didnt buy suv  , wrong predictions of people who didn't buy suv ]
[ wrong predictions of who bought the suv  , correct predictions of  people who bought suv]


```python
accuracy_score(y_test,y_predict)
```

graph (visualising )

```python
len(scale.inverse_transform(x_test))
```

```python
from matplotlib.colors import ListedColormap
x_set,y_set=scale.inverse_transform(x_test),y_test
x1,x2=np.meshgrid(np.arange(start=x_set[:,0].min()-10, stop=x_set[:,0].max() + 10, step=0.25),
                  np.arange(start=x_set[:,1].min()-1000, stop=x_set[:,1].max()+1000, step=0.25))
plt.contourf(x1,x2,classifier.predict(scale.transform(np.array([x1.ravel(),x2.ravel()]).T)).reshape(x1.shape),alpha=0.75,cmap=ListedColormap(('red','green')))
plt.xlim(x1.min(),x1.max())
plt.ylim(x2.min(),x2.max())

for i,j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j,0], x_set[y_set == j,1],c=ListedColormap(('red','green'))(i),label=j)
plt.title('logistic regression (test set)')
plt.xlabel('age')
plt.ylabel("estimated salary")
plt.legend()
plt.show()
```
