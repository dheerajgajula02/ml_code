```python
dataset = pd.read_csv('Mall_Customers.csv')
x=dataset.iloc[:,[3,4]].values
x
```

in order to select only the specific coloums we can follow the syntax

### **using elbow method to find the optimal number of clusters** ###

```python
from sklearn.cluster import KMeans
wcss =[]
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.xlabel('no of clusters')
plt.ylabel('wcss value')
plt.title('elbow method')
plt.show()
```

```python
k_means = KMeans(n_clusters=5,init='k-means++',random_state=42)
y_kmeans=k_means.fit_predict(x)
print(y_kmeans)
```

```python
plt.scatter(x[y_kmeans==0,0],x[y_kmeans==0,1], s=50,c='red', label='cluster 1', edgecolors='#64de84')
plt.scatter(x[y_kmeans==1,0],x[y_kmeans==1,1], s=50, c='blue', label ='cluster 2', edgecolors='#64de84')
plt.scatter(x[y_kmeans==2,0],x[y_kmeans==2,1], s=50,c='green', label='cluster 1', edgecolors='#64de84')
plt.scatter(x[y_kmeans==3,0],x[y_kmeans==3,1], s=50,c='yellow', label='cluster 1', edgecolors='#64de84')
plt.scatter(x[y_kmeans==4,0],x[y_kmeans==4,1], s=50,c='pink', label='cluster 1', edgecolors='#64de84')
plt.scatter(k_means.cluster_centers_[:,0], k_means.cluster_centers_[:,1],s=100,c='#8e4aa1', edgecolors='black')
plt.xlabel('spent')
plt.ylabel('income')
plt.title('k means cluster')
plt.legend()
plt.show()
```
