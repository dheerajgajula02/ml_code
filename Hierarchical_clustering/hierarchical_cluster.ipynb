```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**again here we will use two features coz I want to use graph to visualise**

```python
dataset = pd.read_csv('Mall_Customers.csv')
x=dataset.iloc[:,[3,4]].values
```

### using dendrogram to find the optimal number of clusters ###

```python
import scipy.cluster.hierarchy as sch
gram = sch.dendrogram(sch.linkage(x, method='ward'))
plt.title('dendrogram')
plt.xlabel('customers')
plt.ylabel('euclidian distances')
plt.show()
```

as we are having some clashes between 3 and 5 clusters we will train another model on 3 clusters in next file

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
y_pred =cluster.fit_predict(x)
```

```python
y_pred
```

```python
plt.scatter(x[y_pred==0,0], x[y_pred==0,1], s=50, c='#ed4040', label='cluster 1')
plt.scatter(x[y_pred==1,0], x[y_pred==1,1],s=50, c='#ed40ea', label='cluster 2')
plt.scatter(x[y_pred==2,0], x[y_pred==2,1],s=50, c='#835ae8', label='cluster 3')
plt.scatter(x[y_pred==3,0], x[y_pred==3,1],s=50, c='#308a7e', label='cluster 4')
plt.scatter(x[y_pred==4,0], x[y_pred==4,1],s=50, c='#8a8230', label='cluster 5')
plt.show()
```
