```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
x=dataset.iloc[:,[3,4]].values
```

```python
import scipy.cluster.hierarchy as sch
gram = sch.dendrogram(sch.linkage(x,method='ward'))
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
y_pred = cluster.fit_predict(x)
```

```python
y_pred
```

```python
plt.scatter(x[y_pred==0,0], x[y_pred==0,1], s=50, c='#ed4040', label='cluster 1')
plt.scatter(x[y_pred==1,0], x[y_pred==1,1],s=50, c='#ed40ea', label='cluster 2')
plt.scatter(x[y_pred==2,0], x[y_pred==2,1],s=50, c='#835ae8', label='cluster 3')

plt.show()
```
